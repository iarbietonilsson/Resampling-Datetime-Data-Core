### Imports

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

###  Load Data

````python
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv'
df = pd.read_csv(url)
df
````

## Part 1 Load data & datetime index

### 1. Convert the date column to datetime dtype.

````python
# datetime specifying format Year Month Day
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df['date']
````

### 2. Set the date column as the index.

````python
df = df.set_index('date')
df
````

### 3. Filter out only the required data:

* Keep only data from the year 2000 or later.
* Keep only the following features:
  * precipitation
  * mean_temp
  * min_temp
  * max_temp
  * snow_depth

````python
df = df.loc['2000':, ['precipitation', 'mean_temp', 'min_temp', 
                      'max_temp', 'snow_depth']]

df
````

### 4. Impute any missing values.

````python
# check for null values
df.isna().sum()
````

````python
## interpolate temperature values
df['mean_temp'] = df['mean_temp'].interpolate()
df['max_temp'] = df['max_temp'].interpolate()
df['min_temp'] = df['min_temp'].interpolate()
````

````python
# fill preciptation/snow with 0
df['precipitation'] = df['precipitation'].fillna(0)
df['snow_depth'] = df['snow_depth'].fillna(0)
````

````python
# check for null values again
df.isna().sum()
````

* Resample the data as Monthly frequency.
* Find the date and corresponding value for the max precipitation.
* Plot the feature and time period that corresponds to the question.
  * Make sure to meet the "Visualization Requirements" below!
* Determine the date and value for the most precipitation (the max precipitation).
* Add a vertical line at this date with a label that says the date (using date format: "April 2000") and the max precipitation value.
  * Make sure to add a legend so that your label is visible.
  
#### Q1 Visualization Requirements
Use a combination of pandas, matplotlib tick formatters, tick locators, and the fig.autofmt_xdate method to:
* Create a wide figure (e.g. figsize=(12,3))
* With the title "Precipitation for 2000-2010"
* With a ylabel of "Inches"
* Customize the major xaxis ticks to:
  * Have major xaxis ticks every 1 year.
  * Use 4-digit years for the labels
  * Rotate the major ticks' labels 90 degrees.
* Customize the minor xaxis ticks to:
  * Have minor xaxis ticks every 3 months
  * Do not label the minor ticks.

````python
#import tick customization tools
import matplotlib.ticker as mticks
import matplotlib.dates as mdates
````

````python
## setting the default figsize for this notebook
plt.rcParams['figure.figsize'] = (12,3)
````

````python
#  remake ts
ts = df.copy()
ts
````

````python
# set frequency to monthly
ts_MS_res = ts.asfreq("MS")
````

````python
# resample precipitation monthly
ts_MS_res = ts.resample("MS").mean()
````

````python
ts_MS_res.index
````

````python
ts_MS_res
````

````python
# precipitation df from 2000-2010, monthly
precip_cols = ['precipitation']
ts_MS_res  = ts_MS_res.loc["2000":"2010",precip_cols]
ts_MS_res
````

````python
# find date & max precipitation value
date_max_precip = ts_MS_res['precipitation'].idxmax()
max_precip = ts_MS_res['precipitation'].max()
````

````python
fig, ax = plt.subplots(figsize=(12,3))
ax.plot(ts_MS_res.index, ts_MS_res.values)

ax.set( title='Precipitation for 2000-2010',
      ylabel='Inches')

# customize major ticks 
major_loc =  mdates.YearLocator(1)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)


# customize minor ticks
minor_loc =  mdates.MonthLocator(interval=3)
ax.xaxis.set_minor_locator(minor_loc)

# Making major/minor gridlines visually distinct
ax.grid(which='minor',axis='x',ls=":")
ax.grid(which='major',axis='x',color='k')

ax.axvline(date_max_precip,color='red',lw=3, ls=':',label="Max Precipitation {0} : {1:.2f}".format(date_max_precip.strftime('%b %Y'),max_precip))
ax.legend(bbox_to_anchor = [1, 1]);

## Final xdate tweaks
fig = ax.get_figure()

## rotate the dates
fig.autofmt_xdate(which='major',rotation=90,ha='center')
````

#### Q2 Visualization Requirements
 Use a combination of pandas, matplotlib tick formatters, tick locators, and the fig.autofmt_xdate method to:

* Create a wide figure (e.g. figsize=(12,3))
* With the title "Average Temperature"
* With a ylabel of "Degrees"
* Customize the major xaxis ticks to:
  * Have major xaxis ticks every 5 years.
  * Use 4-digit years for the labels
  * Rotate the major ticks' labels 90 degrees.
* Customize the minor xaxis ticks to:
  * Have minor xaxis ticks every 1 year
  * Do not label the minor ticks.

````python
#  remake ts
ts2 = df.copy()
ts2
````

````python
# set frequency to yearly
ts_A_res = ts2.asfreq("A")
````

````python
# resample precipitation yearly
ts_A_res = ts2.resample("A").mean()
````

````python
ts_A_res.index
````

````python
ts_A_res
````

````python
# avg temp df from 2000-2020, yearly
temp_cols = ['mean_temp']
ts_A_res  = ts_A_res.loc["2000":"2020",temp_cols]
ts_A_res
````

````python
# find date & value for coolest avg temp
date_min_avgtemp = ts_A_res['mean_temp'].idxmin()
min_avgtemp = ts_A_res['mean_temp'].min()
````

````python
fig, ax = plt.subplots(figsize=(12,3))
ax.plot(ts_A_res.index, ts_A_res.values)

ax.set( title='Average Temperature',
      ylabel='Degrees')

# customize major ticks 
major_loc =  mdates.YearLocator(5)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)


# customize minor ticks
minor_loc =  mdates.YearLocator(1)
ax.xaxis.set_minor_locator(minor_loc)

# Making major/minor gridlines visually distinct
ax.grid(which='minor',axis='x',ls=":")
ax.grid(which='major',axis='x',color='k')

ax.axvline(date_min_avgtemp,color='blue',lw=3, ls=':',label="Min Average Temp {0} : {1:.2f}".format(date_min_avgtemp.strftime('%Y'),min_avgtemp))
ax.legend(bbox_to_anchor = [1, 1]);

## Final xdate tweaks
fig = ax.get_figure()

## rotate the dates
fig.autofmt_xdate(which='major',rotation=90,ha='center')
````
